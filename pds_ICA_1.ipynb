Importing the required libraries.

```python
import pandas as pd
import numpy as np
```

Importing dataset (hepatitis dataset) from Github.

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
df.head(10)
```

Creating a new data frame called "numeric_df" where only the columns with numeric values from the original data frame are selected.

```python
numeric_df = df.select_dtypes(include=np.number).copy()
numeric_df
```

Verifying if all the data types present in the new data frame are numericals or not

```python
numeric_df.dtypes
```

Now. lets check if there are any null values present in the data frame.

```python
numeric_df.isna().sum()
```

### Now let's ***replace the null values with the mean values of the respective columns.***

```python
numeric_df.fillna(numeric_df.mean(), inplace=True)
numeric_df
```

Let's verify if the task is performed correctly by checking for null values in the data frame.

```python
numeric_df.isna().sum()
```
