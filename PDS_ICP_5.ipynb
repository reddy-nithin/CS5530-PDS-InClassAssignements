#Importing Required Packages

```python
import pandas as pd
```

```python
data_url="https://raw.githubusercontent.com/JuliaData/CSV.jl/refs/heads/main/test/testfiles/Sacramentorealestatetransactions.csv"
df = pd.read_csv(data_url)
df.head(10)
```

```python
#Find the number of the columns in the data frame
df.shape[1]
```

```python
# Find the mean of the house prices
df.price.mean()
```

```python
# Find the median of the house prices
df.price.median()
```

```python
# Find the different types of property in the dataset
df.type.unique()
```

```python
# Find the maximum price of a "condo" in the dataset
df[df.type == 'Condo'].sort_values('price', ascending=False,).head(1)
```

```python
# Find the maximum price of a "condo" in the dataset
df[df.type == 'Condo']['price'].max()
```

```python
# Find the mean of each type of property in the dataset
df.groupby('type').mean(numeric_only=True)
```
